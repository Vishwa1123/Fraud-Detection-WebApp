```python
# Check available columns
print(data_imported.columns)

# Dropping the Domain column safely
data_modified = data_imported.drop(['Domain'], axis=1, errors='ignore').copy()
```

```python
#checking the data for null or missing values
data_modified.isnull().sum()
```

```python
# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
data_modified = data_modified.sample(frac=1).reset_index(drop=True)
data_modified.head()
```

```python
# Sepratating & assigning features and target columns to X & y
y = data_modified['Label']
X = data_modified.drop('Label',axis=1)
X.shape, y.shape
```

```python
# Splitting the dataset into train and test sets: 80-20 split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12)
X_train.shape, X_test.shape
```

```python
#importing packages
from sklearn.metrics import accuracy_score
```

```python
# Creating holders to store the model performance results
ML_Model = []
accuracy_train = []
accuracy_test = []

#function to call for storing the results
def storeResults(model, a,b):
  ML_Model.append(model)
  accuracy_train.append(round(a, 3))
  accuracy_test.append(round(b, 3))
```

```python
# Decision Tree model 
from sklearn.tree import DecisionTreeClassifier

# instantiate the model 
tree = DecisionTreeClassifier(max_depth = 5)
# fit the model 
tree.fit(X_train, y_train)
```

```python
#predicting the target value from the model for the samples
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
```

```python
#computing the accuracy of the model performance
accuracy_train_tree = accuracy_score(y_train,y_train_tree)
accuracy_test_tree = accuracy_score(y_test,y_test_tree)

print("Decision Tree: Accuracy on training Data: {:.3f}".format(accuracy_train_tree))
print("Decision Tree: Accuracy on test Data: {:.3f}".format(accuracy_test_tree))
```

```python
#checking the feature improtance in the model
plt.figure(figsize=(9,7))
n_features = X_train.shape[1]
plt.barh(range(n_features), tree.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('Decision Tree', accuracy_train_tree, accuracy_test_tree)
```

```python
#importing required packages
import keras
from keras.layers import Input, Dense
from keras import regularizers
import tensorflow as tf
from keras.models import Model
from sklearn import metrics
```

```python
#building autoencoder model

input_dim = X_train.shape[1]
encoding_dim = input_dim  # This might be larger; change as needed

input_layer = Input(shape=(input_dim, ))
# Encoder part
encoder = Dense(encoding_dim, activation="relu",
                activity_regularizer=regularizers.l1(10e-4))(input_layer)
encoder = Dense(int(encoding_dim - 2), activation="relu")(encoder)
encoder = Dense(int(encoding_dim - 4), activation='relu')(encoder)
code = Dense(int(encoding_dim - 6), activation='relu')(encoder)  # Bottleneck

# Decoder part
decoder = Dense(int(encoding_dim - 4), activation='relu')(code)
decoder = Dense(int(encoding_dim - 2), activation='relu')(decoder)
decoder = Dense(input_dim, activation='sigmoid')(decoder)  # Use sigmoid or linear based on your data

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.summary()
```

```python
#compiling the model
autoencoder.compile(optimizer='adam',
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

#Training the model
history = autoencoder.fit(X_train, X_train, epochs=10, batch_size=64, shuffle=True, validation_split=0.2) 
```

```python
# Performance evaluation (using loss, not accuracy)

accuracy_train_auto = autoencoder.evaluate(X_train, X_train)[1]
accuracy_test_auto = autoencoder.evaluate(X_test, X_test)[1]

print('\nAutoencoder: Accuracy on training Data: {:.3f}' .format(accuracy_train_auto))
print('Autoencoder: Accuracy on test Data: {:.3f}' .format(accuracy_test_auto))
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('AutoEncoder', accuracy_train_auto, accuracy_test_auto)
```
